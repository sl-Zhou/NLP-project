```python
def run_scorer(system_output_path, gold_standard_path):
    command = ['python', 'utils/m2scorer/scripts/m2scorer.py', system_output_path, gold_standard_path]
    process = subprocess.run(command, capture_output=True, text=True)
    return process.stdout

# Example usage
results = run_scorer('system_output.txt', 'gold_standard.txt')
print(results)
```

```python
gpt = pd.read_csv('data/final_tmu_gfm_dataset.csv')
```

```python
gpt.columns
```

```python
for i in range(gpt.shape[0]):
    print(f"Original: {gpt.iloc[i]['source']}")
    print(f"Output: {gpt.iloc[i]['output']}")
    print(f"Mistake: {gpt.iloc[i]['Mistakes']}")
    print(f"Explanation of Mistakes: {gpt.iloc[i]['Explanation of Mistakes']}")
    print(f"Corrective Action: {gpt.iloc[i]['Corrective Action']}")
    print("*"*50)
```

```python
filtered_df = gpt[pd.isna(gpt['Mistakes'])]
```

```python
cleaned_df = gpt.dropna(subset=['Mistakes'])
```

```python
cleaned_df.reset_index(inplace=True)
```

```python
del cleaned_df['index']
```

```python
has_nan = cleaned_df.isna().any().any()
```

```python
columns_to_save = ['source', 'output']

cleaned_df[columns_to_save].to_csv('data/clean_tmu.csv', index=False)
```

```python
gpt.to_csv('output.csv', index=False)
```

```python
for column in cleaned_df.columns:
    # Define the file name based on the column name
    file_name = f"{column}.txt"
    
    # Write the column values to a text file, each in a separate line
    with open(file_name, 'w') as file:
        file.write('\n'.join(map(str, cleaned_df[column])))

    print(f"File '{file_name}' created successfully.")
```
